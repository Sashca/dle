```python
import torch
from torchvision import datasets, transforms
from torch import utils
```

```python
data = datasets.MNIST('./test_data', train=True, download=True, transform=transforms.ToTensor())
data_loader = utils.data.DataLoader(data, batch_size=len(data) , shuffle=True)
```

```python
meanGlobal = data.data.float().mean() / 255
stdGlobal = data.data.float().std() / 255
print(meanGlobal)
print(stdGlobal)
```

```python
transform_normalize = transforms.Compose([transforms.ToTensor(), transforms.Normalize((meanGlobal,),(stdGlobal,))])
data_norm = datasets.MNIST('./test_data', train=True, download=True, transform=transform_normalize)
```
